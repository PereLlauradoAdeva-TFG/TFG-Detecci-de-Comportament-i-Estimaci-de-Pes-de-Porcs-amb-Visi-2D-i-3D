### Entrenament dataspace

```python
import os
import torch
import flwr as fl
from ultralytics import YOLO
from collections import OrderedDict

# Configuracio de rutes
RUTA_YAML_ANTIGA = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Antiga/data.yaml'
RUTA_YAML_NOVA = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Nova/data.yaml'
RUTA_DESTI_DRIVE = '/content/drive/MyDrive/Tfg/Dataspaces'

# Model base
MODEL_BASE = '/content/drive/MyDrive/Tfg/DataSpace_Federat/llavor_inicial_small.pt'

# Parametres globals
NUM_RONDES_TOTALS = 20
MIDA_IMATGE = 640

os.makedirs(RUTA_DESTI_DRIVE, exist_ok=True)

# Definicio del client federat
class ClientYOLO(fl.client.NumPyClient):
    def __init__(self, ruta_yaml, nom_granja):
        self.model = YOLO(MODEL_BASE)
        self.ruta_yaml = ruta_yaml
        self.nom_granja = nom_granja

    def get_parameters(self, config):
        return [val.cpu().numpy() for _, val in self.model.model.state_dict().items()]

    def set_parameters(self, parameters):
        claus_model = self.model.model.state_dict().keys()
        params_dict = zip(claus_model, parameters)
        estat_dict = OrderedDict({k: torch.tensor(v) for k, v in params_dict})
        self.model.model.load_state_dict(estat_dict, strict=True)

    def fit(self, parameters, config):
        print(f"\nIniciant entrenament a la granja: {self.nom_granja}")
        self.set_parameters(parameters)

        self.model.train(
            data=self.ruta_yaml,
            epochs=5,
            imgsz=MIDA_IMATGE,
            project=f"runs_federats/{self.nom_granja}",
            name="ronda",
            exist_ok=True,
            verbose=False,
            workers=0
        )
        return self.get_parameters(config={}), 100, {}

    def evaluate(self, parameters, config):
        return 0.0, 10, {"loss": 0.0}

# Estrategia per guardar el model global
class SaveGlobalModelStrategy(fl.server.strategy.FedAvg):
    def aggregate_fit(self, server_round, results, failures):
        aggregated_parameters, aggregated_metrics = super().aggregate_fit(server_round, results, failures)

        if aggregated_parameters is not None:
            print(f"\nGuardant el model global de la ronda {server_round}...")

            tensors_numpy = fl.common.parameters_to_ndarrays(aggregated_parameters)

            model_global = YOLO(MODEL_BASE)
            claus_model = model_global.model.state_dict().keys()
            params_dict = zip(claus_model, tensors_numpy)
            estat_dict = OrderedDict({k: torch.tensor(v) for k, v in params_dict})
            model_global.model.load_state_dict(estat_dict, strict=True)

            nom_fitxer = f'yolo_federat_global_ronda_{server_round}.pt'
            ruta_save_global = os.path.join(RUTA_DESTI_DRIVE, nom_fitxer)

            torch.save({'model': model_global.model}, ruta_save_global)
            print(f"Model de la ronda guardat a: {ruta_save_global}")

        return aggregated_parameters, aggregated_metrics

# Funcio per iniciar el client
def iniciar_client(cid: str) -> fl.client.Client:
    if cid == "0":
        return ClientYOLO(RUTA_YAML_ANTIGA, "Granja_Antiga")
    elif cid == "1":
        return ClientYOLO(RUTA_YAML_NOVA, "Granja_Nova")
    else:
        raise ValueError("ID de client no valid")

estrategia_amb_guardat = SaveGlobalModelStrategy(
    fraction_fit=1.0,
    min_fit_clients=2,
    min_available_clients=2,
)

if __name__ == "__main__":
    print("\nIniciant el servidor central del dataspace federat...")

    if not os.path.exists(RUTA_YAML_ANTIGA) or not os.path.exists(RUTA_YAML_NOVA):
        print("Error: Revisa les rutes. No s'ha trobat el data.yaml.")
    elif not os.path.exists(MODEL_BASE):
         print(f"Error: No s'ha trobat la llavor a {MODEL_BASE}. Executa abans la cel·la corresponent.")
    else:
        fl.simulation.start_simulation(
            client_fn=iniciar_client,
            num_clients=2,
            config=fl.server.ServerConfig(num_rounds=NUM_RONDES_TOTALS),
            strategy=estrategia_amb_guardat,
            client_resources={"num_cpus": 2, "num_gpus": 1}
        )
```

## Avaluació models

```python
from ultralytics import YOLO

print("Iniciant l'avaluacio dels models...\n")

# Datasets de validacio
data_antiga = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Antiga/data.yaml'
data_nova = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Nova/data.yaml'

# Carreguem el model federat
ruta_federat = '/content/drive/MyDrive/Tfg/Dataspaces/yolo_federat_global_ronda_3.pt'
model_federat = YOLO(ruta_federat)

print("Avaluant el model federat a la Granja Antiga...")
resultats_fed_antiga = model_federat.val(data=data_antiga, split='val', verbose=False)
map_fed_antiga = resultats_fed_antiga.box.map50
print(f"Precisio a la Granja Antiga: {map_fed_antiga:.4f}\n")

print("Avaluant el model federat a la Granja Nova...")
resultats_fed_nova = model_federat.val(data=data_nova, split='val', verbose=False)
map_fed_nova = resultats_fed_nova.box.map50
print(f"Precisio a la Granja Nova: {map_fed_nova:.4f}\n")

print("Resum del model federat:")
print(f" - Rendiment Granja Antiga: {map_fed_antiga * 100:.2f}%")
print(f" - Rendiment Granja Nova:   {map_fed_nova * 100:.2f}%")
```

```python
from ultralytics import YOLO
import os

# Rutes dels models locals
RUTA_MODEL_ANTIGA = '/content/drive/MyDrive/Tfg/comparativa_models/yolov8m_entrenament/weights/best.pt'
RUTA_MODEL_NOVA = '/content/drive/MyDrive/Tfg/comparativa_models/yolov8m_definitiu_220imgs/weights/best.pt'

# Rutes dels datasets
YAML_ANTIGA = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Antiga/data.yaml'
YAML_NOVA = '/content/drive/MyDrive/Tfg/Dataspaces/Granja_Nova/data.yaml'

def avaluacio_creuada(ruta_model, nom_model):
    print(f"\nIniciant avaluacio creuada: {nom_model}")

    if not os.path.exists(ruta_model):
        print(f"Error: No s'ha trobat el model a {ruta_model}")
        return

    model = YOLO(ruta_model)

    print(f"Testejant {nom_model} a la Granja Antiga...")
    resultats_antiga = model.val(data=YAML_ANTIGA, split='val', verbose=False)
    map50_antiga = resultats_antiga.box.map50 * 100

    print(f"Testejant {nom_model} a la Granja Nova...")
    resultats_nova = model.val(data=YAML_NOVA, split='val', verbose=False)
    map50_nova = resultats_nova.box.map50 * 100

    print(f"\nResum {nom_model}:")
    print(f" - Precisio Granja Antiga: {map50_antiga:.2f}%")
    print(f" - Precisio Granja Nova:   {map50_nova:.2f}%")

if __name__ == "__main__":
    avaluacio_creuada(RUTA_MODEL_ANTIGA, "Model Local (Granja Antiga)")
    avaluacio_creuada(RUTA_MODEL_NOVA, "Model Local (Granja Nova)")
```

# Codi diagrama de gantt

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
import pandas as pd
from datetime import datetime, timedelta

# Dades exactes seguint els títols de l'usuari i afegint els temps de redacció
data = [
    # SPRINT 1
    {"Task": "1.1 Req. Generals", "Start": "2026-02-09", "End": "2026-03-08", "Sprint": "Sprint 1", "Type": "Task"},
    {"Task": "1.2 Investigació i Configuració Metodologia (Git, Trello...)", "Start": "2026-02-19", "End": "2026-02-26", "Sprint": "Sprint 1", "Type": "Task"},
    {"Task": "1.3 Inf. Inicial", "Start": "2026-02-26", "End": "2026-03-08", "Sprint": "Sprint 1", "Type": "Task"},
    {"Task": "FITA: Informe Inicial", "Start": "2026-03-08", "End": "2026-03-08", "Sprint": "Milestone", "Type": "Milestone"},

    # SPRINT 2
    {"Task": "2.1 Req. Específics veterinaris + IRTA", "Start": "2026-03-09", "End": "2026-03-31", "Sprint": "Sprint 2", "Type": "Task"},
    {"Task": "2.2 Processament Dades", "Start": "2026-03-16", "End": "2026-03-26", "Sprint": "Sprint 2", "Type": "Task"},
    {"Task": "2.3 Inici entrenament CNN i YOLO", "Start": "2026-03-26", "End": "2026-04-10", "Sprint": "Sprint 2", "Type": "Task"},
    {"Task": "2.4 Detecció comportaments bàsics porcs", "Start": "2026-04-10", "End": "2026-05-15", "Sprint": "Sprint 2", "Type": "Task"},
    {"Task": "2.5 Redacció Informe Progrés I", "Start": "2026-04-14", "End": "2026-04-19", "Sprint": "Sprint 2", "Type": "Task"},
    {"Task": "FITA: Informe Progrés I", "Start": "2026-04-19", "End": "2026-04-19", "Sprint": "Milestone", "Type": "Milestone"},

    # SPRINT 3
    {"Task": "3.1 Anàlisis imatges 3D", "Start": "2026-04-20", "End": "2026-04-30", "Sprint": "Sprint 3", "Type": "Task"},
    {"Task": "3.2 Algorisme càlcul pes", "Start": "2026-04-30", "End": "2026-05-10", "Sprint": "Sprint 3", "Type": "Task"},
    {"Task": "3.3 Validació algorisme pes", "Start": "2026-05-10", "End": "2026-05-17", "Sprint": "Sprint 3", "Type": "Task"},
    {"Task": "3.4 Redacció informe progrés II", "Start": "2026-05-12", "End": "2026-05-17", "Sprint": "Sprint 3", "Type": "Task"},
    {"Task": "FITA: Informe progrés II", "Start": "2026-05-17", "End": "2026-05-17", "Sprint": "Milestone", "Type": "Milestone"},

    # SPRINT 4
    {"Task": "4.1 Validació resultats models 2D+3D", "Start": "2026-05-18", "End": "2026-05-28", "Sprint": "Sprint 4", "Type": "Task"},
    {"Task": "4.2 Interfície", "Start": "2026-05-28", "End": "2026-06-07", "Sprint": "Sprint 4", "Type": "Task"},
    {"Task": "FITA: Prop. Final", "Start": "2026-06-07", "End": "2026-06-07", "Sprint": "Milestone", "Type": "Milestone"},
    {"Task": "4.3 Memòria i Codi", "Start": "2026-06-08", "End": "2026-06-21", "Sprint": "Sprint 4", "Type": "Task"},
    {"Task": "FITA: Prop. Presentació", "Start": "2026-06-21", "End": "2026-06-21", "Sprint": "Milestone", "Type": "Milestone"},
    {"Task": "4.4Preparació material entrega", "Start": "2026-06-21", "End": "2026-06-28", "Sprint": "Sprint 4", "Type": "Task"},
    {"Task": "FITA: LLIURAMENT FINAL", "Start": "2026-06-28", "End": "2026-06-28", "Sprint": "Milestone", "Type": "Milestone"},
    {"Task": "FITA: DEFENSA PÚBLICA", "Start": "2026-07-06", "End": "2026-07-06", "Sprint": "Milestone", "Type": "Milestone"},
]

df = pd.DataFrame(data)
df['Start'] = pd.to_datetime(df['Start'])
df['End'] = pd.to_datetime(df['End'])
df['Duration'] = df['End'] - df['Start']

colors = {
    'Sprint 1': '#4f46e5',
    'Sprint 2': '#06b6d4',
    'Sprint 3': '#10b981',
    'Sprint 4': '#f59e0b',
    'Milestone': '#e11d48'
}

fig, ax = plt.subplots(figsize=(15, 11))

for i, row in df.iterrows():
    if row['Type'] == 'Task':
        ax.barh(i, row['Duration'].days, left=row['Start'], color=colors[row['Sprint']], edgecolor='white', linewidth=1.5, height=0.5, zorder=3)
    else:
        ax.scatter(row['Start'], i, marker='o', color=colors['Milestone'], s=130, edgecolor='white', linewidth=2, zorder=4)
        date_str = f"{row['Start'].strftime('%d/%m')}"
        ax.text(row['Start'], i - 0.45, date_str, va='center', ha='center', fontsize=9, fontweight='bold', color=colors['Milestone'])

ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Task'], fontsize=11, color='#1f2937', fontweight='500')
ax.invert_yaxis()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#cbd5e1')
ax.spines['bottom'].set_color('#cbd5e1')

def format_date_cat(x, pos):
    dt = mdates.num2date(x)
    mesos = {2: 'Feb', 3: 'Mar', 4: 'Abr', 5: 'Mai', 6: 'Jun', 7: 'Jul'}
    return f"{dt.day:02d} {mesos.get(dt.month, '')}"

ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_date_cat))
plt.xticks(rotation=0, ha='center', fontsize=10, color='#64748b')

min_date = df['Start'].min() - timedelta(days=5)
max_date = df['Start'].max() + timedelta(days=8)
ax.set_xlim(min_date, max_date)

ax.xaxis.grid(True, linestyle='-', color='#f1f5f9', linewidth=2, zorder=0)
ax.set_axisbelow(True)
ax.set_facecolor('white')

legend_elements = [
    mpatches.Patch(facecolor=colors['Sprint 1'], label='Sprint 1: Bases i Planificació'),
    mpatches.Patch(facecolor=colors['Sprint 2'], label='Sprint 2: Visió 2D'),
    mpatches.Patch(facecolor=colors['Sprint 3'], label='Sprint 3: Integració 3D'),
    mpatches.Patch(facecolor=colors['Sprint 4'], label='Sprint 4: Avaluació i Tancament'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors['Milestone'], markersize=10, label='Fites d\'Avaluació')
]
ax.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1.02, 0.5), title="Llegenda", title_fontproperties={'weight':'bold', 'size':12}, fontsize=11, frameon=False)

plt.title("Diagrama de Gantt", fontsize=16, color='#0f172a', fontweight='bold', pad=20, loc='center')

plt.tight_layout()
plt.savefig("Diagrama_Gantt_Final_Titols.png", dpi=300, bbox_inches='tight')
plt.close()
```
